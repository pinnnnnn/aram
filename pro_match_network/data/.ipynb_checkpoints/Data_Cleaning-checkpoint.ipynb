```python
import json
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Processing

```python
#Read in data
file_path = '/Users/pinnzhang/Desktop/aram/pro_match_network/data/2017_complete.csv'
data_raw = pd.read_csv(file_path, sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
```

```python
selected_columns = ['gameid','league','split', 'patchno','side','position','team','champion']
data = data_raw[selected_columns]

# Subset data by your favorite region, patch, team, etc.
# Here I only subset the data from league = RNG for S7 summer split, 
#  also I drop the team stats from the dataset
rng_data = data[(data['team'] == 'Royal Never Give Up') #Royal Never Give Up, Team SoloMid, Team WE
                & (data['split'] == '2017-W')
                & (data['position'] != 'Team')].reset_index(drop = True)
```

```python
data['team'].unique()
```

```python
rng_data.head(10)
```

```python
champ_pos_count = rng_data[['position', 'champion']].groupby('position').nunique()
champ_pos_count
```

```python
# For each champion, get the number of its appearance for each position
champ_pos_df = rng_data[['position', 'champion']].groupby(['champion', 'position'])\
                                                 .size()\
                                                 .reset_index(name = 'counts')
# Only keeps the position with most appearance
champ_pos_df_max = champ_pos_df.sort_values('counts', ascending=False)\
                               .drop_duplicates('champion')\
                               .sort_index()
champ_pos_df_max['index'] = list(range(len(champ_pos_df_max.index)))
champ_pos_df_max.head()
```

```python
#Initialize an empty graph
G = nx.Graph()
#Add all of the champ names to the graph as node
# with attributes of its corresponding position and number of appearance
for i in range(len(champ_pos_df_max.index)):
    name = champ_pos_df_max.iloc[i,0]
    pos = champ_pos_df_max.iloc[i,1]
    count = champ_pos_df_max.iloc[i,2]
    index = champ_pos_df_max.iloc[i,3]
    G.add_node(name, position = pos, appearance = count, index = index)
```

```python
# For each game played, create a list of champions selected by one team
champion_df = rng_data.groupby(['gameid','team'])['champion'].apply(list)\
             .to_frame()\
             .reset_index()
```

```python
champion_df.head()
```

```python
#Add all of the pair of champion(pair as if both were selected by the same team in a match)
for i in range(len(champion_df.index)):
    champ_list = champion_df.loc[i]['champion']
    temp_list = []
    
    for i in range(5):
        for j in range(i+1, 5):
            temp_list.append((champ_list[i],champ_list[j]))
            
    #Add each pair of champ into the graph
    for pair in temp_list:
        pair_sort = sorted(pair)
        source = pair_sort[0]
        target = pair_sort[1]
        if G.has_edge(source, target):
            # If edge already exists, increment the weight by 1
            G[source][target]['weight'] += 1
        else:
            # Else, create a new edge with weight 1
            G.add_edge(source, target, weight=1)
```

# Creating Network Graph

```python
# Draw the prototype of the network to have a general idea
# The interactive plot will be based on this network
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
nx.draw_networkx(G, ax=ax)
```

```python
# Create the nodes and links into a JSON-readable list
nodes = [{'name': x, 
          'position': G.node[x]['position'], 
          'appearance': int(G.node[x]['appearance']),
         'index': int(G.node[x]['index'])} for x in G.nodes()]

links = [{'source': y[0], 'target': y[1], 'weight' : G[y[0]][y[1]]['weight']} for y in G.edges()]

#create a dictionary with:
# key: champ names
# value: corresponding index
champ_index_dict = {}
for i in range(len(champ_pos_df_max)):
    champ_index_dict[champ_pos_df_max.iloc[i,0]] = champ_pos_df_max.iloc[i,3]
# In links, for source and target, update name with its corresponding index
for link in links:
    link['source'] = int(champ_index_dict[link['source']])
    link['target'] = int(champ_index_dict[link['target']])

# save nodes and links into json file
with open('rng.json', 'w') as f:
    json.dump({'nodes': nodes, 'links': links},
              f, indent=4)
```

```python
%%javascript

// We load the d3.js library from the Web.
require.config({paths:
    {d3: "http://d3js.org/d3.v3.min"}});

require(["d3"], function(d3) {
  // The code in this block is executed when the
  // d3.js library has been loaded.

  // First, we specify the size of the canvas
  // containing the visualization (size of the
  // <div> element).
  var width = 500, height = 500;

  // We create a color scale.
  var color = d3.scale.category10();

  // We create a force-directed dynamic graph layout.
  var force = d3.layout.force()
    .charge(-100)
    .linkDistance(function(d) { return  (1 / d.weight) * 200; })
    .size([width, height]);

  // In the <div> element, we create a <svg> graphic
  // that will contain our interactive visualization.
  var svg = d3.select("#rng_graph").select("svg")
  if (svg.empty()) {
    svg = d3.select("#rng_graph").append("svg")
          .attr("width", width)
          .attr("height", height);
  }

  // We load the JSON file.
  d3.json("rng.json", function(error, graph) {
    // In this block, the file has been loaded
    // and the 'graph' object contains our graph.

    // We load the nodes and links in the
    // force-directed graph.
    force.nodes(graph.nodes)
      .links(graph.links)
      .start();

    // We create a <line> SVG element for each link
    // in the graph.
    var link = svg.selectAll(".link")
      .data(graph.links)
      .enter().append("line")
      .attr("class", "link");

    // We create a <circle> SVG element for each node
    // in the graph, and we specify a few attributes.
    var node = svg.selectAll(".node")
      .data(graph.nodes)
      .enter().append("circle")
      .attr("class", "node")
      .attr("r", function(d) { return Math.sqrt(d.appearance) * 5; })  // radius
      .style("fill", function(d) {
         // The node color depends on the position.
         return color(d.position);
      })
      .call(force.drag);
      
    var label = svg.selectAll(".mytext")
                    .data(graph.nodes)
                    .enter()
                    .append("text")
                    .text(function (d) { return d.name; })
                    .style("text-anchor", "middle")
                    .style("fill", "#555")
                    .style("font-family", "Arial")
                    .style("font-size", 12);
      
    // The name of each node is the node number.
    //node.append("title")
    //    .text(function(d) { return d.name; });
      
    //node.append("text")
    //    .text(function (d) {return d.name});

    // We bind the positions of the SVG elements
    // to the positions of the dynamic force-directed
    // graph, at each time step.
    force.on("tick", function() {
      link.attr("x1", function(d){return d.source.x})
          .attr("y1", function(d){return d.source.y})
          .attr("x2", function(d){return d.target.x})
          .attr("y2", function(d){return d.target.y});

      node.attr("cx", function(d){return d.x})
          .attr("cy", function(d){return d.y});
      
      label.attr("x", function(d){ return d.x; })
          .attr("y", function (d) {return d.y - 10; });

    });
      
    d3.select("#clusterButton").on("click", function () {
        /******************/
        /*  Clustering    */
        /******************/
        netClustering.cluster(graph.nodes, graph.links);
        svg.selectAll(".node").transition().duration(2000).style("fill", function(d) { return color(d.cluster); });
    });
  });
});
```
