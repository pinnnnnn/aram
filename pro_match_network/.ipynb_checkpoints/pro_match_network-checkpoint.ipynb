```python
import json
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Processing

```python
#Read in data
file_path = '/Users/pinnzhang/Desktop/aram/pro_match_network/data/2017_complete.csv'
data_raw = pd.read_csv(file_path, sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
```

```python
selected_columns = ['gameid','league','split', 'patchno','side','position','team','champion']
data = data_raw[selected_columns]

# Subset data by your favorite region, patch, team, etc.
# Here I only subset the data from teamn WE for S7 worlds, 
#  also I drop the team stats from the dataset
we_data = data[(data['team'] == 'Royal Never Give Up') #Royal Never Give Up, Team SoloMid, Team WE
                & (data['split'] == '2017-W')
                & (data['position'] != 'Team')].reset_index(drop = True)
```

```python
data['split'].unique()
```

```python
we_data.head(10)
```

```python
champ_pos_count = we_data[['position', 'champion']].groupby('position').nunique()
champ_pos_count
```

```python
# For each champion, get the number of its appearance for each position
champ_pos_df = we_data[['position', 'champion']].groupby(['champion', 'position'])\
                                                 .size()\
                                                 .reset_index(name = 'counts')
# Only keeps the position with most appearance
champ_pos_df_max = champ_pos_df.sort_values('counts', ascending=False)\
                               .drop_duplicates('champion')\
                               .sort_index()
champ_pos_df_max['index'] = list(range(len(champ_pos_df_max.index)))
champ_pos_df_max.head()
```

```python
#Initialize an empty graph
G = nx.Graph()
#Add all of the champ names to the graph as node
# with attributes of its corresponding position and number of appearance
for i in range(len(champ_pos_df_max.index)):
    name = champ_pos_df_max.iloc[i,0]
    pos = champ_pos_df_max.iloc[i,1]
    count = champ_pos_df_max.iloc[i,2]
    index = champ_pos_df_max.iloc[i,3]
    G.add_node(name, position = pos, appearance = count, index = index)
```

```python
# For each game played, create a list of champions selected by one team
champion_df = we_data.groupby(['gameid','team'])['champion'].apply(list)\
             .to_frame()\
             .reset_index()
```

```python
champion_df.head()
```

```python
#Add all of the pair of champion(pair as if both were selected by the same team in a match)
for i in range(len(champion_df.index)):
    champ_list = champion_df.loc[i]['champion']
    temp_list = []
    
    for i in range(5):
        for j in range(i+1, 5):
            temp_list.append((champ_list[i],champ_list[j]))
            
    #Add each pair of champ into the graph
    for pair in temp_list:
        pair_sort = sorted(pair)
        source = pair_sort[0]
        target = pair_sort[1]
        if G.has_edge(source, target):
            # If edge already exists, increment the weight by 1
            G[source][target]['weight'] += 1
        else:
            # Else, create a new edge with weight 1
            G.add_edge(source, target, weight=1)
```

# Creating Network Graph

```python
# Draw the prototype of the network to have a general idea
# The interactive plot will be based on this network
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
nx.draw_networkx(G, ax=ax)
```

```python
# Create the nodes and links into a JSON-readable list
nodes = [{'name': x, 
          'position': G.node[x]['position'], 
          'appearance': int(G.node[x]['appearance']),
         'index': int(G.node[x]['index'])} for x in G.nodes()]

links = [{'source': y[0], 'target': y[1], 'weight' : G[y[0]][y[1]]['weight']} for y in G.edges()]

#create a dictionary with:
# key: champ names
# value: corresponding index
champ_index_dict = {}
for i in range(len(champ_pos_df_max)):
    champ_index_dict[champ_pos_df_max.iloc[i,0]] = champ_pos_df_max.iloc[i,3]
# In links, for source and target, update name with its corresponding index
for link in links:
    link['source'] = int(champ_index_dict[link['source']])
    link['target'] = int(champ_index_dict[link['target']])

# save nodes and links into json file
with open('we.json', 'w') as f:
    json.dump({'nodes': nodes, 'links': links},
              f, indent=4)
```
